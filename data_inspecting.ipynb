```python
import pandas as pd
import numpy as np
```

```python
DATA_PATH="data/heart.csv"
DATA_PATH_2="data/processed.cleveland.data"
```

```python
cols = [
	'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
	'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target',
]

num_cols = [
	'age', 'trestbps', 'chol', 'thalach', 'oldpeak'
]
cat_cols = [
	'sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'
]
```

```python
df = pd.read_csv(DATA_PATH)
df
```

```python
df2 = df.drop_duplicates().reset_index().drop("index", axis=1)
```

```python
df2 = df2.sort_values('age').reset_index().drop("index", axis=1)
df2
```

```python
cleveland_df = pd.read_csv(
	DATA_PATH_2, header=None, names=df.columns, na_values='?'
)
cleveland_df = cleveland_df.sort_values('age').reset_index().drop("index", axis=1)
cleveland_df
```

```python
na_indices = cleveland_df[cleveland_df.isna().any(axis=1)].index
cleveland_df.iloc[na_indices]
```
